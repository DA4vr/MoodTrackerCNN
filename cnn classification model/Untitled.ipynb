```python
#DATA ALLOCATION.
#make sure the total of the sizes is equal to len(data).If not , adjust accordingly.
#train and val are done during training, test is done during testing 
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)+1
test_size = int(len(data)*.1)+1
```

```python
val_size+train_size+test_size
```

```python
train = data.take(train_size) #take and skip methods from tensorFlow.
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

```python
#either functional or sequential
#sequential is if you have a 1 data model and 1 data output
#functional is multiple models and multiple outputs.
from tensorflow.keras.models import Sequential
#conv2D - convolutional neural network layer 
#MaxPooling2 - condensing layer, goes through your images and actually condenses it down 
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
#establishing the sequential class.
model = Sequential()
```

```python
#1st
#adding layers to the convolution
#an input so whenver youre passing throygh a first layer, it needs to have an input
#16 filters
#3x3 - size of the filters 
#1 stride: number of pixels by which the filter moves over the input image.
#scans over an image and tries to condense or extract the relevant information inside the image for the output classification
#activation relu: takes the image and pass it 
#through a function that looks like this _/, it will convert all negatives to 0 to preserve the positive values
#allows us to take into account non linear patterns  
model.add(Conv2D(16,(3,3),1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
#2nd
model.add(Conv2D(32,(3,3),1, activation='relu'))
model.add(MaxPooling2D())
#3rd
model.add(Conv2D(16,(3,3),1, activation='relu'))
model.add(MaxPooling2D())
#4th
#flatten the data so when we apply convolutional layer , the filters will be the last channel
#so we condense the rows and the width and then the number of filters will form the channel value
#condense it down to a single value
model.add(Flatten())
#5th
#fully connected layers, 256 neurons
model.add(Dense(256, activation ='relu'))
#final layer , single output which will represent 0 or 1 
#sigmoid takes any
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile('adam',loss = tf.losses.BinaryCrossentropy(), metrics= ['accuracy'])
```

```python
model.summary()
```

```python
#TRAIN THE DATA

#create log directory.
logdir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20,validation_data=val, callbacks=[tensorboard_callback]) 
```

```python
#The models performance improved consistently across the epochs.
#The training accuracy started at 42.38% and reached near 100% by the end of the training.
#The validation accuracy also improved significantly, reaching 100% by Epoch 14 and maintaining high accuracy afterward.
#The loss decreased substantially, indicating the model's increasing ability to make accurate predictions.
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color = 'red',label='loss')
plt.plot(hist.history['val_loss'],color='blue',label='val_loss')
fig.suptitle('Loss',fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color = 'purple',label='accuracy')
plt.plot(hist.history['val_accuracy'],color='pink',label='val_accuracy')
fig.suptitle('Accuracy',fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
#EVALUATING PERFORMANCE
```

```python
from tensorflow.keras.metrics import Precision, Recall,BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X,y = batch
    yhat = model.predict(X)
    pre.update_state(y,yhat)
    re.update_state(y,yhat)
    acc.update_state(y,yhat) 
```

```python
print(f'Precision: {pre.result().numpy()}, Recall: {re.result().numpy()}, Accuracy: {acc.result().numpy()}')
```
